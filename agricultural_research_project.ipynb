```python
import pandas as pd

# Complete agriculture dataset - Nsukka 2026
data = {
    "Crop": ["Maize", "Rice", "Bean", "Tomato", "Yam", "Soybean", "Cassava", "Groundnut"],
    "Yield_kg": [1500, 2000, 800, 1200, 3000, 900, 4000, 700],
    "Rainfall_mm": [120, 180, 55, 100, 210, 80, 250, 60],
    "Fertilizer_kg": [50, 80, 30, 45, 100, 35, 120, 25],
    "Price_Naira": [350, 400, 800, 500, 600, 700, 200, 900]
}

df = pd.DataFrame(data)
print("=== AGRICULTURAL DATASET ===")
print(df)
print("\nDataset shape:", df. shape)
```

STATISTICAL SUMMARY

```python
# Statistical summary
print("\n=== STATISTICAL SUMMARY ===")
print(df.describe())

print("\n=== KEY FINDINGS ===")
print("Highest Yielding crop:", df.loc[df['Yield_kg'].idxmax(), 'Crop'])
print("Lowest Rainfall crop:", df.loc[df['Rainfall_mm'].idxmin(), 'Crop'])
print("Average yield:", df['Yield_kg'].mean(), "kg")
print("Rainfall-Yield correlation:",
      round(df["Rainfall_mm"].corr(df["Yield_kg"]), 2))
```

```python
import matplotlib.pyplot as plt
# Create a research visualization dashboard
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Agricultural Research Dashboard - Nsukka Nigeria 2026")

# Chart 1 - Bar chart
axes[0, 0].bar(df['Crop'], df['Yield_kg'], color="green")
axes[0, 0].set_title("Crop Yield Comparison")
axes[0, 0].set_xlabel("Crop")
axes[0, 0].set_ylabel("Yield (kg)")
axes[0, 0].tick_params(axis='x', rotation=45)

# Chart 2 - Scatter plot
axes[0, 1].scatter(df['Rainfall_mm'], df['Yield_kg'], color="blue")
axes[0, 1].set_title("Rainfall vs. Yield Correlation")
axes[0, 1].set_xlabel("Rainfall (mm)")
axes[0, 1].set_ylabel("Yield (kg)")

# Chart 3 - Price bar chart
axes[1, 0].bar(df['Crop'], df['Price_Naira'], color="orange")
axes[1, 0].set_title("Crop Price Comparison")
axes[1, 0].set_xlabel("Crop")
axes[1, 0].set_ylabel("Price (Naira)")
axes[1, 0].tick_params(axis='x', rotation=45)

# Chart 4 - Pie chart
axes[1, 1].pie(df['Yield_kg'], labels=df['Crop'], autopct='%1.1f%%')
axes[1, 1].set_title("Crop Yield Distribution")

plt.tight_layout()

```

```python
# Save complete dataset
df.to_csv("agricultural_research_nsukka_2026.csv", index=False)
print("Research data saved successfully!")
print("Total crop analyzed:", len(df))
print("Research location: Nsukka, Nigeria")
print("Year: 2026")
```
